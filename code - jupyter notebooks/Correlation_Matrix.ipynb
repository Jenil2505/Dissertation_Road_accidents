```python
import  numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the dataset
df = pd.read_csv('dft-road-casualty-statistics-collision-last-5-years (1).csv')
#df
```

```python
df.columns
```

```python
df.head()
```

```python
# Compute the correlation matrix
corr_matrix = df.corr()
#corr_matrix
# Plot the correlation matrix
plt.figure(figsize=(20, 20))
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5, mask=mask)
plt.title('Correlation Matrix')
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

selected_columns = ['speed_limit', 'road_type', 'first_road_class', 'first_road_number','junction_detail','second_road_class','number_of_casualties']
df_selected = df[selected_columns]

# Calculate the correlation matrix using the Spearman method
corr_matrix = df_selected.corr(method='spearman')
# Plot the correlation matrix
plt.figure(figsize=(7,7))
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5, vmin=-1, vmax=1)
plt.title('Correlation Matrix: Speed limit vs road_type, first_road_class, first_road_number, junction_detail, second_road_class, number_of_casualties')
plt.show()
```
