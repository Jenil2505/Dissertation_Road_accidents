```python
import numpy as np
import pandas as pd
df = pd.read_csv("RTA-2018-2022.csv")
import statsmodels.formula.api as smf


model= smf.mnlogit('accident_severity~C(weather_conditions)+C(road_surface_conditions)+C(light_conditions)+C(road_type)+C(junction_detail)+C(carriageway_hazards)', df).fit()
model.summary()
```
