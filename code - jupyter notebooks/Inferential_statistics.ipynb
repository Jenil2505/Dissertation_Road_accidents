```python
import pandas as pd
import scipy.stats as stats
from scipy.stats import chi2

df = pd.read_csv("RTA-2018-2022.csv")
print("\n")
df = df[df['light_conditions'] != -1] 
 
crosstab = pd.crosstab(df["accident_severity"], df["light_conditions"])
print(crosstab)
print("\n")
 
w, x, y, z = stats.chi2_contingency(crosstab)
print("The Chi Square value is:", w)
print("The P-value is:", x)
print("The value for degree of freedom is :", y)
print("Expected cell counts is:", z)
print("\n")
 
```

```python
import pandas as pd
import scipy.stats as stats
 
df = pd.read_csv("RTA-2018-2022.csv")
print("\n")
df = df[df['weather_conditions'] != -1]
 
crosstab = pd.crosstab(df["accident_severity"], df["weather_conditions"])
print(crosstab)
print("\n")
 
w, x, y, z = stats.chi2_contingency(crosstab)
print("The Chi Square value is:", w)
print("The p-value is:", x)
print("The value for degree of freedom is :", y)
print("Expected cell counts is:", z)
print("\n")

alpha = 0.05  # significance level

if x < alpha:
    print("Reject the null hypothesis. There is a significant relationship between the variables.")
else:
    print("Fail to reject the null hypothesis. There is no significant relationship between the variables.")
```

```python
#########################EXTRA WORK
import pandas as pd
import scipy.stats as stats

# Read in the dataset
data = pd.read_csv("RTA-2018-2022.csv")

weather_conditions = data["weather_conditions"]
light_conditions = data["light_conditions"]
accident_severity = data["accident_severity"]

# Create a contingency table
crosstab = pd.crosstab(index=[weather_conditions, light_conditions], columns=accident_severity)
print(crosstab)

# Perform Chi-Square test of independence
chi2, p, dof, expected = stats.chi2_contingency(crosstab)

# Print results
print("Chi-Square Statistic:", chi2)
print("p-value:", p)
print("Degrees of Freedom:", dof)
print("Expected Frequencies:")
print(expected)
```

```python
import pandas as pd
import scipy.stats as stats

data = pd.read_csv("RTA-CASUALTY.csv")

group1 = data[data['sex_of_casualty'] == 1]['age_band_of_casualty']
group2 = data[data['sex_of_casualty'] == 2]['age_band_of_casualty']

t_statistic, p_value = stats.ttest_ind(group1, group2, equal_var=False)  # Assuming unequal variances

# Print results
print("T-Statistic:", t_statistic)
print("P-Value:", p_value)
if p_value < 0.05:
    print("Reject the null hypothesis, Statistically significant difference exists between groups.")
else:
    print("Fail to reject the null hypothesis, No statistically significant difference exists between groups.")
```

```python
import pandas as pd
import scipy.stats as stats

data = pd.read_csv("dft-road-casualty-statistics-vehicle-last-5-years.csv")

group1 = data[data['sex_of_driver'] == 1]['age_band_of_driver']
group2 = data[data['sex_of_driver'] == 2]['age_band_of_driver']

t_statistic, p_value = stats.ttest_ind(group1, group2, equal_var=False)  # Assuming unequal variances

# Print results
print("T-Statistic:", t_statistic)
print("P-Value:", p_value)
if p_value < 0.05:
    print("Reject the null hypothesis, Statistically significant difference exists between groups.")
else:
    print("Fail to reject the null hypothesis, No statistically significant difference exists between groups.")
```
